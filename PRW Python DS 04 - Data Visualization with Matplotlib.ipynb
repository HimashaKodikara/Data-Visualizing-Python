# Plotting multiple charts together

```python
x=np.arange(10)
y1=x+5
y2=x**2
plt.plot(x,y1,"b*",x,y2,"go")
plt.title("Two plots together")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=np.arange(10)
y1=x+5
y2=x**2
plt.plot(x,y1,"b-",y2,"r-")
plt.title("Two plots together")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Subplots in the same figure

# First method

```python
x=np.arange(10)
y1=x+5
y2=x**2
plt.figure(figsize=(10,5))

#Plots are created with a single row and two columns
plt.subplot(1,2,1)
plt.plot(x,y1,"b*")
plt.title("First chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(1,2,2)
plt.plot(x,y2,"go")
plt.title("Second chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
x=np.arange(10)
y1=x+5
y2=x**2
y3=-x
y4=np.sqrt(x)
plt.figure(figsize=(10,10))

#Plots are created with a two rows and two columns
plt.subplot(2,2,1)
plt.plot(x,y1,"b*")
plt.title("First chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2,2,2)
plt.plot(x,y2,"go")
plt.title("Second chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2,2,3)
plt.plot(x,y3,"r+")
plt.title("Third chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2,2,4)
plt.plot(x,y4,"k-")
plt.title("Fourth chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

# Second method

```python
x=np.arange(10)
y1=-(x**2)
y2=x**3
y3=np.exp(x)
y4=np.sqrt(x)

#Plots are created with a two rows and two columns
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,10))

ax[0,0].plot(x,y1,"b*")
ax[0,0].set_title("First chart")
ax[0,0].set_xlabel("X")
ax[0,0].set_ylabel("Y")

ax[0,1].plot(x,y2,"go")
ax[0,1].set_title("Second chart")
ax[0,1].set_xlabel("X")
ax[0,1].set_ylabel("Y")

ax[1,0].plot(x,y3,"r+")
ax[1,0].set_title("Thrird chart")
ax[1,0].set_xlabel("X")
ax[1,0].set_ylabel("Y")

ax[1,1].plot(x,y4,"k-")
ax[1,1].set_title("Fourth chart")
ax[1,1].set_xlabel("X")
ax[1,1].set_ylabel("Y")


plt.show()
```
